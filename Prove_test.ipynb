```python
#Ottenere i dati rGGRUPPATI PER regione, primi 10 paesi per confermati (last update, confermati, morti, ricoverati)

df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-17-2020.csv")

ord_df =df.groupby(['Country/Region'])["Confirmed", "Recovered"].sum().sort_values(['Confirmed'], ascending = False).head(10)
ord_df



mask1 = ord_df["Recovered"].between(1000, 5000)
mask2 = ord_df["Confirmed"] >  1000
final_df = ord_df[mask1 & mask2]
#len(ord_df[mask1 & mask2])
final_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

final_df.plot(kind = "bar", y = 'Confirmed', ylabel='Number of confirmed cases')


#df.plot(kind ='bar', x='name', y='punteggio', xlabel='name', ylabel='punteggio')
```
